```python
1#triangulo

lado1 = int(input("insira o valor do primeiro lado: "))
lado2 = int(input("insira o valor do segundo lado: "))
lado3 = int(input("insira o valor do terceiro lado: "))

triangulo = lado1 + lado2 + lado3

if lado1 == lado2 and lado2 == lado3:
  print("O seu triangulo com perimeto de: ", triangulo, "é Equilatero")
else:
  if lado1 == lado2 and lado1 != lado3 or lado1 != lado2 and lado1 == lado3:
   print("O seu triangulo com perimeto de: ", triangulo, "é Isóceles")
  else:
    print("O seu triangulo com perimeto de: ", triangulo, "é escaleno")
```
