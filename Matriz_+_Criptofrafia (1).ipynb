# Multiplicação de Matrizes

````python
import numpy as np

A = np.array([[1, 2],[3, 4],])
B = np.array([[-2, 1],[1.5,-0.5]])
print('A ')
print(A)
print('\nB ')
print(B)
I = np.dot(A, B) # AxB - Se o produto de AxB é uma matriz identidade, B é inverso de A
print('\nIdentidade: ')
print(I)
````

Matriz identidade - elemento neutro da multiplicação de Matrizes:

````python
A2 = np.dot(A, I)
print('A x I :')
print(np.dot(A,I)) # AxI
print('\nB x I :')
print(np.dot(B,I)) # BxI
````
