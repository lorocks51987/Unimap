**Exercício 4: Verificação de Triângulo**

Solicite ao usuário que insira os três lados de um triângulo. Verifique e imprima se é um triângulo equilátero, isósceles ou escaleno.

```python
1#triangulo

lado1 = int(input("insira o valor do primeiro lado: "))
lado2 = int(input("insira o valor do segundo lado: "))
lado3 = int(input("insira o valor do terceiro lado: "))

triangulo = lado1 + lado2 + lado3

if lado1 == lado2 and lado2 == lado3:
  print("O seu triangulo com perimeto de: ", triangulo, "é Equilatero")
else:
  if lado1 == lado2 and lado1 != lado3 or lado1 != lado2 and lado1 == lado3:
   print("O seu triangulo com perimeto de: ", triangulo, "é Isóceles")
  else:
    print("O seu triangulo com perimeto de: ", triangulo, "é escaleno")
```

Crie um script que solicite ao usuário que insira números inteiros positivos. Continue somando os números até que o usuário insira um número negativo. Imprima a soma final.

```python
x = int(input("Digite o valor: "))
soma = 0

while x >= 0:
    soma += x
    x = int(input("Digite o valor: "))
print("A soma dos valores digitados é:", soma)
```

Desenvolva um script que peça ao usuário para inserir uma palavra. Verifique se a palavra é um palíndromo (lê-se igual de trás para frente). Imprima o resultado.

```python
palavra = input("Digite a palavra:").lower()

while palavra != palavra[::-1]:
    palavra = input("A palavra não é um palíndromo. Digite novamente:")

print("A palavra é um palíndromo.")
```

**Exercício 6: Verificação de Números Primos**

Peça ao usuário para inserir um número. Verifique se o número é primo ou não e imprima o resultado. Lembre-se que um número primo é aquele que só é divisível por 1 e por ele mesmo.

```python

num = int(input("Digite o numero: "))

primo = True
i = 2

while 2 < num:
  if num % i == 0:
    primo = False
  i += 1

if primo:
  print("O número é primo.")
else:
  print("O número não é primo.")
```

Crie um script que solicite ao usuário inserir as notas de um aluno em quatro disciplinas. Utilize um loop `for` para calcular a média das notas e imprima o resultado.

**Exercício 8: Tabuada Personalizada**

Peça ao usuário para inserir um número. Utilize um loop `for` para imprimir a tabuada desse número de 1 a 10.

```python

valor = int(input("digite o valor: "))
for i in range(1,11):

  print(valor, "X", i, "=", valor*i)
```
