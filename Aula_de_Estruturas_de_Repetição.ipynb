7) Os números primos tem grande aplicação em funções de Criptografia. Um número primo é aquele
que é divisível apenas por um e por ele mesmo. Faça um programa em Python que peça um número
inteiro e determine se ele é ou não um número primo.

```python
num = int(input("Digite o Numero: "))

i = 2
noprimo=False
while (i<num and noprimo==False):
  if(num%i)==0:
    print("Não é primo")
    noprimo=True
  i=i+1
if(noprimo==False):
  print("é primo")
```

8) Altere o programa de cálculo dos números primos, informando, caso o número não seja primo, por
quais número ele é divisível sem resto.


```python

num = int(input("Digite o Numero: "))

i = 2
noprimo=False
divisores=[]
while (i<num and noprimo==False):
  if(num%i)==0:
    divisores.append(i)
    noprimo=True
  i=i+1
if(noprimo==False):
  print("é primo")
else:
  print("Não é primo")
  print("Divisores:",divisores)
```

11) Faça um programa em Python que peça para n pessoas a sua idade, ao final o programa deverá
verificar se a média de idade da turma varia entre 0 e 25, 26 e 60 e maior que 60; e então, dizer se a
turma é jovem, adulta ou idosa, conforme a média calculada.

```python
n = int(input("Quantas pessoas na turma? "))
soma_idades = 0
for i in range(n):
  idade = int(input(f"Digite a idade da pessoa {i+1}: "))
  soma_idades += idade

media_idade = soma_idades / n

if media_idade <= 25:
    print("A turma é jovem.")
elif media_idade <= 60:
    print("A turma é adulta.")
else:
    print("A turma é idosa.")
```

12) Numa eleição existem três candidatos (A, B e C). Faça um programa que peça o número total de
eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
eleitores = int(input("Quantidade de eleitores: "))
votos_a = 0
votos_b = 0
votos_c = 0

for i in range(eleitores):
  voto = input("Escolha seu candidato (A, B ou C): ").upper()
  if voto == "A":
    votos_a += 1
  elif voto == "B":
    votos_b += 1
  elif voto == "C":
    votos_c += 1
  else:
    print("Voto inválido.")

print("O candidato A teve:", votos_a, "votos.")
print("O candidato B teve:", votos_b, "votos.")
print("O candidato C teve:", votos_c, "votos.")
```

13) Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade
de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.


```python
turmas = int(input("Quantidade de turmas: "))
total_alunos = 0

for i in range(turmas):
  alunos = int(input(f"Quantidade de alunos da turma {i+1}: "))
  if alunos > 40:
    print("A turma não pode ter mais de 40 alunos.")
  else:
    total_alunos += alunos

media_alunos = total_alunos / turmas

print(f"O número médio de alunos por turma é: {media_alunos:.2f}")
```

14) O Departamento Estadual de Meteorologia lhe contratou para desenvolver um programa que leia
um conjunto indeterminado de temperaturas (até que seja informado o número 9999), e ao final
informar a menor e a maior temperatura e a média das temperaturas.


```python
menor_temperatura = float('inf')
maior_temperatura = float('-inf')
soma_temperaturas = 0
quantidade_temperaturas = 0

while True:
    temperatura = float(input('Digite a temperatura (ou 9999 para sair): '))
    if temperatura == 9999:
        break
    if temperatura < menor_temperatura:
        menor_temperatura = temperatura
    if temperatura > maior_temperatura:
        maior_temperatura = temperatura
    soma_temperaturas += temperatura
    quantidade_temperaturas += 1

media_temperaturas = soma_temperaturas / quantidade_temperaturas

print('Menor temperatura:', menor_temperatura)
print('Maior temperatura:', maior_temperatura)
print('Média das temperaturas:', media_temperaturas)
```

15) Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de
crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de
1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população
do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.


```python
populacao_a = 80000
taxa_crescimento_a = 3/100
populacao_b = 200000
taxa_crescimento_b = 1.5/100
anos = 0

while populacao_a < populacao_b:
    populacao_a += populacao_a * taxa_crescimento_a
    populacao_b += populacao_b * taxa_crescimento_b
    anos += 1

print(f"A população do país A ultrapassará ou igualará a população do país B em {anos} anos.")
```
