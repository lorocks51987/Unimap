```python
clientes = {}

# Função para adicionar um cliente
nome = input("Digite o nome do cliente: ")
email = input("Digite o email do cliente: ")
if email not in clientes:
    clientes[email] = {"nome": nome, "pontos": 0}
    print(f"Cliente {nome} adicionado com sucesso!")
else:
    print("Este cliente já está registrado.")

# Simulando uma compra
email_cliente = input("Digite o email do cliente para fazer a compra: ")
if email_cliente in clientes:
    valor_compra = float(input("Digite o valor da compra: "))
    pontos_ganhos = valor_compra // 10
    clientes[email_cliente]["pontos"] += pontos_ganhos
    print(f"Compra realizada! O cliente ganhou {pontos_ganhos} pontos.")
else:
    print("Cliente não encontrado.")

# Simulando um resgate de pontos
email_cliente = input("Digite o email do cliente para resgatar pontos: ")
if email_cliente in clientes:
    pontos = int(input("Digite a quantidade de pontos a ser resgatada: "))
    if pontos <= clientes[email_cliente]["pontos"]:
        desconto = pontos * 0.1
        clientes[email_cliente]["pontos"] -= pontos
        print(f"Você resgatou {pontos} pontos e obteve R${desconto:.2f} de cashback.")
    else:
        print("Pontos insuficientes para resgate.")
else:
    print("Cliente não encontrado.")
```
